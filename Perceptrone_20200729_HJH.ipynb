```python
def init_newtwort():
    network = {}
    network["weight1"] = weight1 = np.array([[0.1, 0.3, 0.5], [0.2, 0.4, 0.6]])
    network["bias1"] = bias1 = np.array([0.1, 0.2, 0.3])
    network["weight2"] = weight2 = np.array([[0.1, 0.4], [0.2, 0.5], [0.3, 0.6]])
    network["bias2"] = bias2 = np.array([0.1, 0.2])
    network["weight3"] = weight3 = np.array([[0.1, 0.3], [0.2, 0.4]])
    network["bias3"] = bias3 = np.array([0.1, 0.2])
    
    return network
```

```python
def forward(network, x):
    weight1, weight2, weight3 = network["weight1"], network["weight2"], network["weight3"]
    bias1, bias2, bias3 = network["bias1"], network["bias2"], network["bias3"]
    
    result_of_inner1 = np.dot(x, weight1) + bias1
    output1 = sigmoid(result_of_inner1)
    result_of_inner2 = np.dot(output1, weight2) + bias2
    output2 = sigmoid(result_of_inner2)
    result_of_inner3 = np.dot(output2, weight3) + bias3
    y = identity_function(result_of_inner3)
    
    return y
```

```python
network = init_newtwort()
x = np.array([1.0, 0.5])
y = forward(network, x)
print(y)
```
